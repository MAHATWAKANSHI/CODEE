```python
# Import the h2o package, which is an open-source platform for machine learning and AI
import h2o

# Initialize the H2O cluster, which sets up the connection to the H2O server
h2o.init()
```

```python
# Import the pandas package
import pandas as pd

# Import the Document class from the langchain.docstore.document module
# This class is likely used to represent and work with documents in the langchain framework
from langchain.docstore.document import Document
```

```python
# Load the CSV file into a pandas dataframe
df = pd.read_csv('formal.csv')

# This was a format i used... you can alter as per your requirement
# Combine the question and answer columns into a single text column
df['text'] = df['Student Greeting'] + ' ' + df['Teacher Response']
```

```python
# Create a list of documents with a 'page_content' attribute
# This list comprehension iterates over each text item in the 'text' column of the DataFrame 'df'
# Each text item is used to create a Document object with 'page_content' set to the text
documents = [Document(page_content=text) for text in df['text'].tolist()]

# Import the RecursiveCharacterTextSplitter class from the langchain.text_splitter module
from langchain.text_splitter import RecursiveCharacterTextSplitter

# Create an instance of RecursiveCharacterTextSplitter with a chunk size of 500 characters and an overlap of 50 characters
text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)

# Split the documents into smaller text chunks based on the specified chunk size and overlap
text_chunks = text_splitter.split_documents(documents)
```

```python
# Import the HuggingFaceEmbeddings class from the langchain.embeddings module
from langchain.embeddings import HuggingFaceEmbeddings

# Import the FAISS class from the langchain.vectorstores module
from langchain.vectorstores import FAISS

# Load the embedding model from Hugging Face
# Use the 'sentence-transformers/all-MiniLM-L6-v2' model for generating sentence embeddings
# Set the model to run on the CPU
embeddings = HuggingFaceEmbeddings(model_name='sentence-transformers/all-MiniLM-L6-v2', model_kwargs={'device': 'cuda:0'})

# Convert the text chunks into embeddings and create a FAISS vector store
# The FAISS vector store will store the embeddings and allow for efficient similarity search
vector_store = FAISS.from_documents(text_chunks, embeddings)
```

```python
# Import the hf_hub_download function from the huggingface_hub module
from huggingface_hub import hf_hub_download

# Define the repository ID and model basename for the model to be downloaded from Hugging Face Hub
repo_id = "h2oai/h2o-danube2-1.8b-chat-GGUF"  # The ID of the repository on Hugging Face Hub
model_basename = "h2o-danube2-1.8b-chat-Q8_0.gguf"  # The specific filename of the model within the repository

# Download the model file from Hugging Face Hub using the specified repository ID and model filename
model_path = hf_hub_download(repo_id=repo_id, filename=model_basename)  # This function returns the local file path to the downloaded model
```

```python
''''# Import the CTransformers class from the langchain.llms module
from langchain.llms import CTransformers

# Initialize the CTransformers language model
# Use the model file downloaded from Hugging Face Hub (model_path)
# Specify the model type as "gpt"
# Set additional configuration parameters:
# - 'max_new_tokens': the maximum number of tokens to generate in a response
# - 'temperature': controls the randomness of the model's output (lower values make the output more deterministic)
llm = CTransformers(model=model_path, model_type="gpt", config={'max_new_tokens': 72, 'temperature': 0.06})'''

from langchain.llms import CTransformers

# Initialize the CTransformers language model with additional parameters for consistency
llm = CTransformers(
    model=model_path,  # Path to your model file
    model_type="gpt",  # Choose the model type
    config={
        'max_new_tokens': 30,  # Maximum number of tokens to generate
        'temperature': 0.01,    # Lower temperature for less randomness
        'top_p': 0.5,          # Nucleus sampling to keep output focused
        'top_k': 50            # Limit to top k tokens for selection
    }
)

```

```python
# Import the PromptTemplate class from the langchain.prompts module
from langchain.prompts import PromptTemplate

# Define the QA prompt template with a more precise style (Change accordingly as per your need)
template = """
Let's find the answer together! Here is some context that might help:
{context}
Now, based on this context, here is the precise answer to your question:
Question: {question}
Answer:
"""

# Create a PromptTemplate instance with the defined template
# The template expects two input variables: 'question' and 'context'
qa_prompt = PromptTemplate(template=template, input_variables=['question', 'context'])

# Implement a post-processing function to ensure the response is clean
def clean_response(response):
    # Remove any text after an unintended question or irrelevant information
    stop_phrases = ["Question:", "prompt:"]  # List of phrases to stop at
    for phrase in stop_phrases:
        if phrase in response:
            response = response.split(phrase)[0]  # Split response at the stop phrase and take the first part
    return response.strip()  # Remove leading and trailing whitespace

# Example function to generate a response
def generate_response(question, context):
    # Format the prompt using the QA prompt template with the given question and context
    prompt = qa_prompt.format(question=question, context=context)
    # Ensure the model is only generating up to the answer
    response = llm.generate(prompt, max_tokens=500)  # Generate response with a maximum of 100 tokens
    return clean_response(response)  # Clean and return the response
```

```python
# Import the RetrievalQA class from the langchain.chains module
from langchain.chains import RetrievalQA

# Update the RetrievalQA chain with the new prompt template
# Create a RetrievalQA instance using the from_chain_type method
# Parameters:
# - llm: the language model instance (initialized CTransformers in this case)
# - chain_type: type of chain ('stuff' in this example)
# - retriever: vector_store converted to a retriever with search parameters {'k': 2}
# - return_source_documents: flag to indicate if source documents should be returned along with the answer
# - chain_type_kwargs: additional keyword arguments specific to 'stuff', including the prompt and document variable name

chain = RetrievalQA.from_chain_type(
    llm=llm,  # Language model instance
    chain_type='stuff',  # Type of chain being used
    retriever=vector_store.as_retriever(search_kwargs={'k': 2}),  # Vector store converted to a retriever with search parameters
    return_source_documents=False,  # Flag to return source documents along with the answer
    chain_type_kwargs={'prompt': qa_prompt, 'document_variable_name': 'context'}  # Additional kwargs for 'stuff', including prompt and document variable name
)
```

```python
from gtts import gTTS
from IPython.display import Audio, display
import re

# Function to speak text
def speak(text):
    tts = gTTS(text=text, lang='en')
    tts.save('output.mp3')
    display(Audio('output.mp3', autoplay=True))

def punctuation(text):
    # Use regex to find the last punctuation mark
    match = re.search(r'[.!?]', text)
    if match:
        # Cut the text up to and including the last punctuation mark
        text = text[:match.end()]
    return text

# Main loop to keep the chat going
while True:
    user_message = input("Student: ")
    if user_message.lower() in ["exit", "quit"]:
        print("Thank You. Goodbye!")
        break
    response = chain.invoke({'query': user_message})
    print('Teacher:',punctuation(response['result']))
    speak(punctuation(response['result']))
```
